# IF statements

##### IF and Else Statements

```python
number=int(input("Enter a Number: "))
if number%2 == 0:
    print("The number is even")
```

```python
number=int(input("Enter a Number: "))
if number%2 == 0:
    print("The number is even")
else:
    print("The number is odd")
```

```python
a=int(input("Enter a Number: "))
b=int(input("Enter a Number: "))
if a>b:
    print("A is greater than B")
elif a<b:
    print("A is Less than B")
elif a==b:
    print("A is equal to B")
else:
    print("Nothing")
```

## Line plot

```python
plt.figure(figsize=(20,5))
plt.plot(df.sepal_length,color="yellow")
plt.xlabel("Sepal_length")
plt.ylabel("Measurement")
plt.title("Line plot for sepal Length",color="red")
plt.show()
```

## Boxplot

```python
plt.boxplot(df.sepal_width)
plt.show()
```

> In sepal width we are having the outliers

```python
plt.bar(df.species,df.sepal_width)
plt.show()
```

## Bar chart

```python
plt.barh(df.species,df.sepal_width)
plt.show()
```

## Histogram

```python
plt.hist(df.petal_length)
plt.show()
```

## Pie Chart

```python
values=df.species.value_counts()
```

```python
names=["setosa","versicolor","virginica"]
```

```python
plt.pie(values,labels=names,autopct="%1.4f%%",colors=["red","green","yellow"],explode=[0,0,0.3])
plt.title("The distribution of Species")
plt.show()
```
